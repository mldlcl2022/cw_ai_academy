# Resampling

## Down Sampling
기존 데이터의 샘플링 개수를 감소시키는 것

```python
data = pd.DataFrame({'날짜' : ['2023-05-01','2023-05-02','2023-05-03'],
                     '판매량' : [10,15,12]})
data.info()
```

```python
data['날짜'] = pd.to_datetime(data['날짜'])
data.info()
```

```python
data = data.set_index('날짜')
data
```

```python
data_downs = data.resample('W').sum()
data_downs
```

```python
data_downs = data.resample('W-MON').sum()
data_downs
```

## Up Sampling
기존 데이터의 샘플링 개수를 증가시키는 것

```python
data_ups = data.resample('H').sum()
data_ups
```

# Interpolation

```python
data = pd.DataFrame({'date' : pd.date_range(start = '2023-05-01', end = '2023-05-14', freq = 'D'),
                     'value' : [1,6,4,None,36,17,None,91,180,23,88,12,44,23]})
data.info()
```

```python
data = data.set_index('date')
data.head(3)
```

```python
data_downs = data.resample('2D').mean()
data_downs
```

```python
data_interpolation = data.interpolate(method = 'linear')
data_interpolation
```

# Rolling Window
이동평균, 추세 같은 것을 확인할 때 활용 가능

```python
data = pd.DataFrame({'date' : pd.date_range(start = '2023-05-01', end = '2023-05-10', freq = 'D'),
                     'value' : [1,6,4,12,36,17,44,91,180,23]})
data
```

```python
data = data.set_index('date')
data
```

```python
data_rolling = data['value'].rolling(window = 5).mean()
data_rolling
```

```python
data_rolling = data['value'].rolling(window = 5)

for i in data_rolling :
    print(i,'\n')
```
