# Interpolation

```python
df = pd.DataFrame({'A' : [1,2,3,None,5],
                   'B' : [None,20,30,None,50]})
df
```

```python
df = df.interpolate()
df
```

```python
df = pd.DataFrame({'A' : [1,2,3,None,5],
                   'B' : [None,20,30,None,50]})
df = df.interpolate(method = 'nearest')
df
```

```python
df = pd.DataFrame({'A' : [1,2,3,None,5],
                   'B' : [None,20,30,None,50]})
df = df.interpolate(method = 'quadratic')
df
```

```python
df = pd.DataFrame({'A' : [1,2,3,None,5],
                   'B' : [None,20,30,None,50]})
df = df.interpolate(method = 'nearest').bfill()
df
```

```python
df = pd.DataFrame({'A' : [1,2,3,None,5],
                   'B' : [None,20,30,None,50]})
df = df.interpolate(method = 'nearest').dropna()
df
```

# Concat

```python
df1 = pd.DataFrame({'고객 ID' : ['1','2','3'],
                    '이름' : ['John','Mike','Sarah'],
                    '나이' : [25,35,30]})

df2 = pd.DataFrame({'고객 ID' : ['4','5','6'],
                    '이름' : ['Adam','Kim','Park'],
                    '나이' : [222,17,32]})
```

```python
df_concat = pd.concat([df1, df2]).reset_index(drop = True)
df_concat
```

# Merge

```python
df1 = pd.DataFrame({'고객 ID' : ['1','2','3'],
                    '이름' : ['John','Mike','Sarah'],
                    '나이' : [25,35,30]})

df2 = pd.DataFrame({'고객 ID' : ['4','5','6'],
                    '이름' : ['Adam','Kim','Park'],
                    '나이' : [222,17,32]})
```

```python
df_merge = pd.merge(df1, df2, how = 'outer')
df_merge
```

# Boolean filtering

```python
data = pd.DataFrame({'이름' : ['아리','노아','미아','리오'],
                     '나이' : [3,5,4,2],
                     '성별' : ['여자','남자','여자','남자'],
                     '키' : [110,120,115,105]})
data
```

```python
import sklearn
from sklearn.datasets import load_iris

raw_data = load_iris()
data = pd.DataFrame(data = raw_data.data, columns = raw_data.feature_names)

data['Target'] = raw_data.target
data.head(3)
```

```python
def custom_summary(column) :
    s = {'mean' : column.mean(),
         'median' : column.median(),
         'std' : column.std(),
         'max' : column.max(),
         'min' : column.min()}
    return pd.Series(s)
```

```python
data.apply(custom_summary)
```

```python
data.describe()
```

# Group by

```python
data = pd.DataFrame({'City' : ['Seoul','Seoul','Busan','Busan','Incheon'],
                     'Year' : [2020,2021,2020,2021,2020],
                     'Population' : [9.7,9.8,3.4,3.6,2.9]})
data
```

```python
data.groupby('City')
```

```python
for i in data.groupby('City') :
    print(i)
```

```python
print('City별 Population')
data.groupby('City')['Population'].mean()
```

### 예제 1

### 예제 3

```python
data = pd.DataFrame({'City' : ['Seoul','Seoul','Busan','Busan','Incheon'],
                     'Population' : [9.7,9.8,3.4,3.6,2.9],
                     'GDP' : [620,650,180,200,150],
                     'Area' : [605.2,605.2,770.1,770.1,1049]})
data
```

```python
def  max_min(column) :
    return column.max() - column.min()
```

```python
data.groupby('City')[['Population','GDP']].apply(max_min)
```
